```python
# List of numerical variables
numerical_vars = ['AGE', 'SMOKING', 'YELLOW_FINGERS', 'ANXIETY', 'PEER_PRESSURE', 
                  'CHRONIC_DISEASE', 'FATIGUE', 'ALLERGY', 'WHEEZING', 'ALCOHOL_CONSUMING', 
                  'COUGHING', 'SHORTNESS_OF_BREATH', 'SWALLOWING_DIFFICULTY', 'CHEST_PAIN']
# Plot the relationship between numerical variables and LUNG_CANCER
plt.figure(figsize=(15, 20))
for i, var in enumerate(numerical_vars):
    plt.subplot(5, 3, i + 1)
    sns.boxplot(x='LUNG_CANCER', y=var, data=df)
    plt.title(f'{var} vs LUNG_CANCER')
plt.tight_layout()
plt.show()
```

```python
# Calculate correlation matrix for numerical variables
correlation_matrix = df.corr()

# Plot heatmap of the correlation matrix
plt.figure(figsize=(15, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Matrix')
plt.show()
```

```python
from sklearn.preprocessing import LabelEncoder

# Encode categorical variables
df_encoded = df.copy()
df_encoded['GENDER'] = LabelEncoder().fit_transform(df['GENDER'])
df_encoded['LUNG_CANCER'] = LabelEncoder().fit_transform(df['LUNG_CANCER'])

# Calculate correlation matrix including encoded categorical variables
correlation_matrix_encoded = df_encoded.corr()

# Plot heatmap
plt.figure(figsize=(14, 12))
sns.heatmap(correlation_matrix_encoded, annot=True, cmap='Oranges', linewidths=0.5, fmt='.2f')
plt.title('Correlation Matrix with Encoded Categorical Variables')
plt.show()
```

```python
# Insights 
"""
LUNG CANCER have negative correlation with AGE and WHEEZING
CHEST PAIN is correlated with PEER PRESSURE 
SHORTNESS OF BREATH and COUGHING correlated to ALLERGY
ALCOHOL CONSUMPTION is correlated to SWALLOWING DIFFICULTY 
SMOKING is correlated to ANXIETY and CHRONIC DISEASE
"""
```

```python
# Filter data to include only smokers
smokers = df[df['SMOKING'] == 1]

# Count how many smokers have and do not have lung cancer
counts = smokers['LUNG_CANCER'].value_counts()

# Plot pie chart with percentages
plt.figure(figsize=(8, 8))
plt.pie(counts, labels=counts.index, autopct='%1.1f%%', colors=['#ff9999','#66b3ff'], startangle=140)
plt.title('Proportion of Smokers with and without Lung Cancer')
plt.show()
```

```python
# Filter smokers older than 50
smokers_older_than_50 = df[(df['SMOKING'] == 1) & (df['AGE'] > 50)]

# Count how many of these smokers have and do not have lung cancer
counts = smokers_older_than_50['LUNG_CANCER'].value_counts()

# Calculate percentages
percentages = (counts / counts.sum()) * 100

# Create a DataFrame for plotting
plot_data = pd.DataFrame({'Count': counts, 'Percentage': percentages})

# Plot bar plot with percentages
plt.figure(figsize=(10, 6))
bar_plot = sns.barplot(x=plot_data.index, y=plot_data['Count'], palette='viridis')

# Annotate bars with percentages
for i, p in enumerate(bar_plot.patches):
    height = p.get_height()
    label = f'{plot_data["Percentage"].iloc[i]:.1f}%'
    bar_plot.annotate(label,
                      (p.get_x() + p.get_width() / 2., height),
                      ha='center', va='center',
                      xytext=(0, 9),
                      textcoords='offset points')
plt.xlabel('LUNG_CANCER')
plt.ylabel('Count')
plt.title('Number of Smokers Over 50 with and without Lung Cancer (Percentages Included)')
plt.show()
```
