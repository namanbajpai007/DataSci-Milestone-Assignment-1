```python
# As job_type and company_type are categorical in nature we will fill the missing values using mode.
df['job_type'] = df['job_type'].fillna(df['job_type'].mode()[0])
df['company_type'] = df['company_type'].fillna(df['company_type'].mode()[0])
```

```python
# Replace non-English entries with their English equivalents
df['company_type'] = df['company_type'].replace({
    'Сфера услуг, сервиса, бытового обслуживания Включает в том числе: Прачечная/химчистка, ремонт мебели, часов, ювелирны...': 'service industry'
})

# Verify the replacement
print(df['company_type'].unique())
```

```python
# Distribution of company types 
plt.figure(figsize=(14, 8))
sns.countplot(data=df, y='company_type', palette='husl', order=df['company_type'].value_counts().index)
plt.title('Distribution of Company Types')
plt.xlabel('Count')
plt.ylabel('Company Type')
plt.show()
# Maximum companies are of Manufacturing industary 
```

```python
# Generate the pie chart
job_type_counts = df['job_type'].value_counts()

# Explode parameter to separate slices
explode = [0.1] * len(job_type_counts)  # Adjust the value to separate slices

plt.figure(figsize=(10, 7))
plt.pie(
    job_type_counts,
    labels=job_type_counts.index,
    autopct='%1.1f%%',
    colors=plt.cm.tab20.colors,
    shadow=True,  # Adds shadow for 3D effect
    explode=explode  # Separates slices
)
plt.title('Distribution of Job Types')
plt.show()
# Most of them work in commercial organizations 
```

```python
# Outlier Detection 

plt.figure(figsize=(14, 6))

# Box plot for age
plt.subplot(1, 2, 1)
sns.boxplot(x=df['age'])
plt.title('Box Plot of Age')

# Box plot for weight1
plt.subplot(1, 2, 2)
sns.boxplot(x=df['weight1'])
plt.title('Box Plot of Weight1')

plt.show()
```

```python
from scipy import stats

# Calculate Z-scores
df['age_zscore'] = stats.zscore(df['age'])
df['weight1_zscore'] = stats.zscore(df['weight1'])

# Identify outliers
age_outliers = df[(df['age_zscore'] > 3) | (df['age_zscore'] < -3)]
weight1_outliers = df[(df['weight1_zscore'] > 3) | (df['weight1_zscore'] < -3)]

print("Age Outliers:\n", age_outliers)
print("Weight1 Outliers:\n", weight1_outliers)
```

```python
plt.figure(figsize=(14, 6))

# Box plot for age with outliers highlighted
plt.subplot(1, 2, 1)
sns.boxplot(x=df['age'], palette='Set3')
sns.scatterplot(data=age_outliers, x='age', y=[1]*len(age_outliers), color='red', label='Outliers')
plt.title('Box Plot of Age with Outliers')

# Box plot for weight1 with outliers highlighted
plt.subplot(1, 2, 2)
sns.boxplot(x=df['weight1'], palette='Set3')
sns.scatterplot(data=weight1_outliers, x='weight1', y=[1]*len(weight1_outliers), color='red', label='Outliers')
plt.title('Box Plot of Weight1 with Outliers')

plt.legend()
plt.show()

# Weight1 contains Outliers 
```

```python
# Set the figure size
plt.figure(figsize=(14, 6))

# Violin plot for age
plt.subplot(1, 2, 1)
sns.violinplot(x='age', data=df, palette='dark')
plt.title('Violin Plot of Age')

# Violin plot for weight1
plt.subplot(1, 2, 2)
sns.violinplot(x='weight1', data=df, palette='Set1')
plt.title('Violin Plot of Weight1')

plt.show()
```

```python
# Calculate IQR for weight1
Q1_weight1 = df['weight1'].quantile(0.25)
Q3_weight1 = df['weight1'].quantile(0.75)
IQR_weight1 = Q3_weight1 - Q1_weight1

# Define bounds for outliers
lower_bound = Q1_weight1 - 1.5 * IQR_weight1
upper_bound = Q3_weight1 + 1.5 * IQR_weight1

# Identify outliers
outliers = df[(df['weight1'] < lower_bound) | (df['weight1'] > upper_bound)]
print(f"Number of outliers in weight1: {len(outliers)}")

# Remove outliers
df_no_outliers = df[(df['weight1'] >= lower_bound) & (df['weight1'] <= upper_bound)]

# Plot the violin plot again without outliers
plt.figure(figsize=(10, 6))
sns.violinplot(x='weight1', data=df_no_outliers, palette='pastel')
plt.title('Violin Plot of Weight1 (Without Outliers)')
plt.show()
```

```python
# Bivariate Analysis
# Categorical vs Categorical
plt.figure(figsize=(12, 6))
sns.heatmap(pd.crosstab(df['sex'], df['will_vote']), annot=True, cmap='coolwarm', fmt='d')
plt.title('Heatmap of Gender vs Will Vote')
plt.xlabel('Will Vote')
plt.ylabel('Gender')
plt.show()
# Female voters are more than male voters 
```

```python
# Bivariate Analysis

# 1. Education vs Income
education_income_crosstab = pd.crosstab(df['education'], df['income'])
plt.figure(figsize=(10, 6))
sns.heatmap(education_income_crosstab, annot=True, cmap='YlGnBu', fmt='d')
plt.title('Heatmap of Education vs Income')
plt.xlabel('Income')
plt.ylabel('Education')
plt.show()

plt.figure(figsize=(12, 6))
education_income_crosstab.plot(kind='bar', stacked=True, colormap='Paired')
plt.title('Stacked Bar Plot of Education by Income')
plt.xlabel('Education')
plt.ylabel('Count')
plt.legend(title='Income')
plt.show()
```

```python
# 2. Employment vs Income
employment_income_crosstab = pd.crosstab(df['employment'], df['income'])
plt.figure(figsize=(10, 6))
sns.heatmap(employment_income_crosstab, annot=True, cmap='YlGnBu', fmt='d')
plt.title('Heatmap of Employment vs Income')
plt.xlabel('Income')
plt.ylabel('Employment')
plt.show()

plt.figure(figsize=(12, 6))
employment_income_crosstab.plot(kind='bar', stacked=True, colormap='Paired')
plt.title('Stacked Bar Plot of Employment by Income')
plt.xlabel('Employment')
plt.ylabel('Count')
plt.legend(title='Income')
plt.show()
```

```python
# 3. Education vs Employment
education_employment_crosstab = pd.crosstab(df['education'], df['employment'])
plt.figure(figsize=(10, 6))
sns.heatmap(education_employment_crosstab, annot=True, cmap='YlGnBu', fmt='d')
plt.title('Heatmap of Education vs Employment')
plt.xlabel('Employment')
plt.ylabel('Education')
plt.show()

plt.figure(figsize=(12, 6))
education_employment_crosstab.plot(kind='bar', stacked=True, colormap='Paired')
plt.title('Stacked Bar Plot of Education by Employment')
plt.xlabel('Education')
plt.ylabel('Count')
plt.legend(title='Employment')
plt.show()
```

```python
# Filter the data to include only those who will vote
df_will_vote = df[df['will_vote'] == 'definitely yes']

# Calculate the vote counts for each candidate
vote_counts = df_will_vote['candidate'].value_counts()

# Check if the filtered dataframe or vote_counts is empty
if df_will_vote.empty:
    print("No data available for those who will vote.")
else:
    # Plot
    plt.figure(figsize=(10, 6))
    ax = sns.barplot(x=vote_counts.index, y=vote_counts.values, palette='viridis')

    # Add percentage labels on the bars
    total_votes = vote_counts.sum()
    for p in ax.patches:
        percentage = f'{(p.get_height() / total_votes) * 100:.1f}%'
        ax.annotate(percentage, 
                    (p.get_x() + p.get_width() / 2., p.get_height()), 
                    ha='center', va='center', 
                    xytext=(0, 10), textcoords='offset points')

    plt.title('Distribution of Votes Among Candidates')
    plt.xlabel('Candidate')
    plt.ylabel('Number of Votes')
    plt.show()
# Putin is winning the election with very great margin
```
